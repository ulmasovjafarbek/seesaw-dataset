Scope 1 emissions are direct greenhouse (GHG) emissions that occur from sources that are controlled or owned by an organization (e.g., emissions associated with fuel combustion in boilers, furnaces, vehicles). Scope 2 emissions are indirect GHG emissions associated with the purchase of electricity, steam, heat, or cooling. 

take pres = 0.5 for calculation

```python
import pandas as pd

df = pd.read_csv("2013_-_Global_500_Emissions_and_Response_Status.csv")
```

```python
df
```

```python
df2 = df[["Company Name ", "Ticker Symbol ","Scope 1 (metric tonnes CO2e)", "Scope 2 (metric tonnes CO2e)"]].copy()
df2
```

```python
df3=df2.dropna()
df3
```

```python
df3["scope 1 + scope 2"] = df3.apply(lambda x: x["Scope 1 (metric tonnes CO2e)"]+x["Scope 2 (metric tonnes CO2e)"], axis=1)
df3
```

```python
def normalize_z(df):
    dfout = df.copy()
    dfout = (df - df.mean(axis=0)) / df.std(axis=0)
    return dfout
```

```python
df3["sustainability index"] = 100-((df3["scope 1 + scope 2"]-df3["scope 1 + scope 2"].min(axis=0))/(df3["scope 1 + scope 2"].max(axis=0)-df3["scope 1 + scope 2"].min(axis=0))*100)
```

```python
df3
```

```python
def clean_symbol(symbol):
    l = symbol.split()
    return l[0]

df3["Symbol"] = df3.apply(lambda x: clean_symbol(x["Ticker Symbol "]), axis=1)
df3
```

```python
df_sus = df3[["Company Name ", "Symbol","sustainability index"]]
df_sus
```

```python
stock_df = pd.read_csv("stock.csv")
stock_df
```

```python
def clean_price(price):
    return price[1::]

stock_df["Price"] = stock_df.apply(lambda x: clean_price(x["Last Sale"]), axis=1)
stock_df
```

```python
def clean_beta(change_percent):
    try:
        num = (change_percent)[:-1]
        return float(num) * 0.01
    except:
        return None

stock_df["Beta"] = stock_df.apply(lambda x: clean_beta(x["% Change"]), axis=1)
```

```python
stock_df_info = stock_df[["Symbol", "Name", "Market Cap", "Price", "Beta"]]
stock_df_info
```

```python
stock_df_info.dropna()
```

```python
df_sus_stock=df_sus.merge(stock_df_info, how="inner", on="Symbol")
```

```python
df_sus_stock
```

```python
def cal_Ri(beta, r_market):
    return 0.035+float(beta)*(float(r_market)-0.035)

df_sus_stock["R_i"] = df_sus_stock.apply(lambda x: cal_Ri(x["Beta"], x["Price"]),axis=1)
df_sus_stock
```

```python
sigma_square = df_sus_stock["Price"].apply(lambda x: float(x)).var()
sigma_square
```

```python
import numpy as np
pres = 0.5
def cal_signal(pres, sigma_square, R_i):
    return pres*np.log(sigma_square)+(2-pres)*R_i

df_sus_stock["signal"] = df_sus_stock.apply(lambda x: cal_signal(pres, sigma_square, x["R_i"]), axis=1)
df_sus_stock
```

```python
df_sus_stock["profitability index"] = ((df_sus_stock["signal"]-df_sus_stock["signal"].min(axis=0))/(df_sus_stock["signal"].max(axis=0)-df_sus_stock["signal"].min(axis=0)))*100
```

```python
df_sus_stock
```

```python
result_df = df_sus_stock[["Company Name ", "Symbol", "Price","Market Cap","sustainability index", "profitability index"]]
```

```python
result_df
```

```python
desired_s = 96
desired_p = 90

import math
def cal_similarity(company_s, company_p):
    return math.sqrt((company_s-desired_s)**2+(company_p-desired_p)**2)


result_df["absolute distance"] = result_df.apply(lambda x: cal_similarity(x["sustainability index"],x["profitability index"]), axis=1)
result_df
```

```python
# now what i am doing is to make the absolute distance to be in between 0 to 100 scale
min_dis = result_df["absolute distance"].min(axis=0)
max_dis = result_df["absolute distance"].max(axis=0)
result_df["result similarity"] = 100 - ((result_df["absolute distance"]-min_dis)/(max_dis-min_dis))*100
result_df
```

```python
result_df.to_csv('result.csv', index=False)
```
